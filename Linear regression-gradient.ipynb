```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import inv
from numpy.linalg import norm
```

```python
df = pd.read_csv("../data/Data-Train.csv")
```

```python
def NormalizeData(data):
    return (data - np.mean(data)) / (np.std(data))

def computeCost(y_p, y):

    m = len(y)
    inner = np.power(( y_p - y), 2)
    return np.sum(inner) / (m)



def gradient_descent(X, y, theta, iteration, alpha):

    past_costs = []
    past_thetas = [theta]
    for i in range(iteration):

        prediction = np.dot(X, theta)
        cost = computeCost(prediction,y)
        past_costs.append(cost)
        gradients = (1 / m) * X.T.dot(X.dot(theta) - y)
        theta = theta - (alpha * gradients)
        past_thetas.append(theta)

        diff = abs(past_thetas[i] - past_thetas[i-1])
        if np.sum(diff) < precision:
            print("Converged iteration: ",i)
            break

    return past_thetas, past_costs
```

```python
my_Data = np.array(df)
m = np.shape(my_Data)[0]
X = np.array([np.ones(m),my_Data[:, 0]]).T
y = np.array(my_Data[:, 1]).T
y = y.reshape(-1, 1)


alpha = 0.0001
iteration = 10000
theta = np.zeros(2)
theta = theta.reshape(-1, 1)
precision = 0.00001


past_thetas, past_costs = gradient_descent(X, y, theta, iteration, alpha)
newtheta = past_thetas[-1]
print("Theta Values:", newtheta[0], newtheta[1])
print("MSE error for trian: ",past_costs[-1])

plt.title('Cost Function')
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
past_costs = np.array(past_costs)
plt.plot(past_costs.reshape(-1))
plt.savefig("../images/LR gradient1.jpg")
plt.show()

y_hat = np.dot(X,newtheta)
plt.plot(X[:,1], y,"b.",label = "Training data")
plt.plot(X[:,1],y_hat, "r--",label = "regression line")
plt.savefig("../images/LR gradient2.jpg")
plt.legend()

df_test =pd.read_csv("../data/Data-Test.csv")
my_Data_test = np.array(df_test)

m_test = np.shape(my_Data_test)[0]
X_test = np.array([np.ones(m_test), my_Data_test[:,0]]).T
y_test = np.array(my_Data_test[:,1]).T
y_test = y_test.reshape(-1,1)


y_hat_test = np.dot(X_test,newtheta)
print("MSE error for test: ",computeCost(y_hat_test,y_test))

plt.figure()
plt.plot(X_test[:,1], y_test, "b.",label = "Test data")
plt.plot(X_test[:,1], y_hat_test, "r--",label = "Regression line")
plt.legend()
plt.savefig("../images/LR gradient3.jpg")
plt.show()
```
