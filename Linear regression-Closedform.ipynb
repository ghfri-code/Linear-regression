```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
df = pd.read_csv("/content/drive/MyDrive/Portfolio-Codes/pattern recegnition/Linear Regresion/data/Data-Train.csv")
```

```python
def computeCost(y_p, y):

    m = len(y)
    inner = np.power(( y_p - y), 2)

    return np.sum(inner) / (m)

def best_param(X, y):

    X_transpose = X.T
    print((X_transpose.dot(X)))
    best_params = inv(X_transpose.dot(X)).dot(X_transpose).dot(y)

    return best_params
```

```python
my_Data = np.array(df)
m = np.shape(my_Data)[0]

X = np.array([np.ones(m), my_Data[:,0]]).T
y = np.array(my_Data[:,1]).T

theta = best_param(X,y)
print("theta values : ",theta)
y_hat = np.dot(X,theta)

plt.plot(X[:,1],y, "b.",label = "trainig data")
plt.plot(X[:,1],y_hat, "r--",label = "regression line")
plt.savefig("/content/drive/MyDrive/Portfolio-Codes/pattern recegnition/Linear Regresion/images/LR closedform1.jpg")
plt.legend()

print("MSE error for train: ",computeCost(y_hat,y))


df_test =pd.read_csv("/content/drive/MyDrive/Portfolio-Codes/pattern recegnition/Linear Regresion/data/Data-Test.csv")
my_Data_test = np.array(df_test)

m_test = np.shape(my_Data_test)[0]
X_test = np.array([np.ones(m_test), my_Data_test[:,0]]).T
y_test = np.array(my_Data_test[:,1]).T



prediction = np.dot(X_test,theta)


plt.figure()
plt.plot(X_test[:,1], y_test, "b.",label = "test data")
plt.plot(X_test[:,1], prediction, "r--",label = "regression line")
plt.legend()
plt.savefig("/content/drive/MyDrive/Portfolio-Codes/pattern recegnition/Linear Regresion/images/LR closedform2.jpg")
print("MSE error for test: ",computeCost(prediction,y_test))

plt.show()
```
